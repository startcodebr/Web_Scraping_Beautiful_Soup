```
import bs4 as bs
import urllib.request

source = urllib.request.urlopen('https://www.saude.mg.gov.br/component/search/?all=&exact=covid-19&any=&none=&created=&modified=&area=stories').read()
```

```
soup = bs.BeautifulSoup(source,'lxml')
```

```
# title of the page
print(soup.title)

# get attributes:
print(soup.title.name)

# get values:
print(soup.title.string)

# beginning navigation:
print(soup.title.parent.name)

# getting specific values:
print(soup.p)
```

```
print(soup.find_all('p'))
```

```
for paragraph in soup.find_all('p'):
    print(paragraph.string)
    print(str(paragraph.text))
```

```
for url in soup.find_all('a'):
    print(url.get('href'))
```

```
nav = soup.nav
for url in nav.find_all('a'):
    print(url.get('href'))
```

```
body = soup.body
for paragraph in body.find_all('p'):
    print(paragraph.text)
```

```
for div in soup.find_all('div', class_='content'):
    print(div.text)
```

```
#scrapping specifically with a table example
table = soup.table

#find the table rows within the table
table_rows = table.find_all('tr')

# iterate through the rows,find the td tags, and then print out each of the table data tags:
for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    print(row)
```

```
soup = bs.BeautifulSoup(source,'lxml')
for div in soup.find_all('div', class_='content'):
  for url in div.find_all('a'):
      print(url.get('href'))
```

```
soup = bs.BeautifulSoup(source,'lxml')
for div in soup.find_all('div', class_='content'):
  for url in div.find_all('a'):
      print(url.text)
```

```
soup = bs.BeautifulSoup(source,'xml')
for url in soup.find_all('loc'):
    print(url.text)
```

```
soup = bs.BeautifulSoup(source,'lxml')

js_test = soup.find('p', class_='jstest')

print(js_test.text)
```
