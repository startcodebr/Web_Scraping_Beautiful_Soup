```
import requests
website_url = requests.get('https://pt.wikipedia.org/wiki/Lista_de_redes_sociais').text
```

```
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
#print(soup.prettify())
```

```
My_table = soup.find('table',{'class':'wikitable sortable'})
My_table
```

```
links = My_table.findAll('a')
links
```

```
redesociais = []
for link in links:
    if link.get('title') != None:
      redesociais.append(link.get('title'))
    
print(redesociais)
```

```
import pandas as pd
df = pd.DataFrame()
df['Redes Sociais'] = redesociais

df
```
